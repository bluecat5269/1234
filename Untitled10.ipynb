```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
import webbrowser
from ipywidgets import interact
import sympy as sp
from sklearn.preprocessing import MinMaxScaler
import qgrid
from matplotlib.font_manager import FontProperties
```

```python
demo=pd.read_csv('123.csv')
```

```python
qgrid.set_grid_option('maxVisibleRows', 12) ### 使用qgrid 把表格可控化
demo=demo.dropna()
q = qgrid.show_grid(demo)
```

```python
q
```

```python
str_choice = "K"
loc=demo[demo['K'].str.contains(str_choice, na=False)]
loc
```

```python
plt.figure(figsize=(15,10),dpi=100,linewidth = 2)
plt.plot(loc.K,loc.A_insert,'s-',color = 'r', label="第一個input array 產生方法")
plt.plot(loc.K,loc.B_insert,'o-',color = 'g', label="第二個input array 產生方法")
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei']
plt.title("Insertion_sort所需時間", x=0.5, y=1.03,fontsize=20)
plt.xticks(fontsize=15)
plt.xlabel("array length", fontsize=30, labelpad = 15)
plt.yticks(fontsize=15)
plt.ylabel("time", fontsize=30, labelpad = 15)
plt.legend(loc = "best", fontsize=20)
```
